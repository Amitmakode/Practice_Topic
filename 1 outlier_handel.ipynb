```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate a dataset with 99 random elements and one outlier
dataset = np.concatenate([np.random.normal(0, 1, 99), [10]])

# Visualize the dataset before treating the outlier
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.title("Original Dataset")
plt.boxplot(dataset)
plt.xlabel("Dataset")
plt.ylabel("Values")

# Find and treat the outlier
outlier_index = np.argmax(dataset)
outlier_value = dataset[outlier_index]

# Replace the outlier with the median value
dataset[outlier_index] = np.median(dataset)

# Visualize the dataset after treating the outlier
plt.subplot(1, 2, 2)
plt.title("Dataset after treating outlier")
plt.boxplot(dataset)
plt.xlabel("Dataset")
plt.ylabel("Values")

# Apply log transform
log_transformed = np.log1p(dataset)

# Apply box transform
box_transformed, _ = np.boxcox(dataset)

# Visualize the transformed datasets
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.title("Log Transform")
plt.hist(log_transformed, bins=20, edgecolor='black')
plt.xlabel("Log Transformed Values")
plt.ylabel("Frequency")

plt.subplot(1, 3, 2)
plt.title("Box Transform")
plt.hist(box_transformed, bins=20, edgecolor='black')
plt.xlabel("Box Transformed Values")
plt.ylabel("Frequency")

plt.subplot(1, 3, 3)
plt.title("Original Dataset")
plt.hist(dataset, bins=20, edgecolor='black')
plt.xlabel("Original Values")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
